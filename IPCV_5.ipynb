```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
# Load image
img = cv2.imread('/content/drive/My Drive/IPCV LAB/img2.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
img
```

```python
# Sobel X (vertical edges)
sobel_x = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=5)

# Sobel Y (horizontal edges)
sobel_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=5)
```

```python
sobel_combined = cv2.magnitude(sobel_x, sobel_y)

# Convert to 8-bit
sobel_combined = cv2.convertScaleAbs(sobel_combined)
```

```python
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1,2,2)
plt.imshow(sobel_combined, cmap='gray')
plt.title('Sobel Edge Detection')
plt.axis('off')

plt.show()
```

```python
import cv2
```

```python
img = cv2.imread('/content/drive/My Drive/IPCV LAB/pexels-tkirkgoz-16063191.jpg')

if img is None:
    print("Error: Image not loaded. Please check the path and file existence.")
else:
    print("Image loaded successfully.")
```

```python
# 2. Convert to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
# 3. Apply Gaussian blur to reduce noise
blurred = cv2.GaussianBlur(gray, (5, 5), 0)
```

```python
# 4. Apply Canny edge detection
# Thresholds 50 and 150 are examples; adjust based on the image
edges = cv2.Canny(blurred, 50, 150)
```

```python
from google.colab.patches import cv2_imshow

# 5. Display the result
cv2_imshow(edges)
```

```python
# Resize the image for better viewing if it's too large
# Calculate the aspect ratio and new dimensions
height, width = edges.shape
max_width = 800
if width > max_width:
    ratio = max_width / width
    new_width = max_width
    new_height = int(height * ratio)
    resized_edges = cv2.resize(edges, (new_width, new_height), interpolation=cv2.INTER_AREA)
else:
    resized_edges = edges # No resizing needed if already small enough

print(f"Original dimensions: {width}x{height}")
print(f"Resized dimensions: {resized_edges.shape[1]}x{resized_edges.shape[0]}")

cv2_imshow(resized_edges)
```

```python
# 1. Read the image in grayscale
img = cv2.imread('/content/drive/My Drive/IPCV LAB/adorable-cat-lifestyle.jpg', cv2.IMREAD_GRAYSCALE)

if img is None:
    print("Error: Image not loaded. Please check the path and file existence.")
else:
    print("Image loaded successfully.")
```

```python
# 2. Apply Gaussian Blur to reduce noise sensitivity
blurred = cv2.GaussianBlur(img, (3, 3), 0)
```

```python
# 3. Apply Laplacian operator
# CV_64F allows for negative numbers (white-to-black edges)
laplacian = cv2.Laplacian(blurred, cv2.CV_64F, ksize=3)
```

```python
# 4. Convert back to absolute unit8 to visualize
abs_laplacian = cv2.convertScaleAbs(laplacian)
```

```python
from google.colab.patches import cv2_imshow

# 5. Display results
cv2_imshow(img)
cv2_imshow(abs_laplacian)
```
